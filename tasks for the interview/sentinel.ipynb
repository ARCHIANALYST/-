```python
import pandas as pd
df = pd.read_excel(r'C:\Users\Arthur\Documents\Data\sentinel\Тестовое_задание.xlsx', sheet_name = 'Корректировка_данных')
df.info()
```

```python
df['balance_bucket'].unique()
```

```python
df.groupby('balance_bucket').agg({'BALLANCE' :'max'})
# 50 тыс. и менее
```

```python
df.groupby('balance_bucket').agg({'BALLANCE' :'min'})
# 101-300 тыс
```

```python
df.loc[(df['BALLANCE'] <= 50999) & (df['balance_bucket'].isna()), 'balance_bucket'] = '50 тыс. и менее'
df.loc[(df['BALLANCE'] >= 51000) & (df['BALLANCE'] <= 100000) & (df['balance_bucket'].isna()), 'balance_bucket'] = '51-100 тыс.'
df.loc[(df['BALLANCE'] >= 100001) & (df['BALLANCE'] <= 300999) & (df['balance_bucket'].isna()), 'balance_bucket'] = '101-300 тыс.'
df.loc[(df['BALLANCE'] >= 301000) & (df['BALLANCE'] <= 500999) & (df['balance_bucket'].isna()), 'balance_bucket'] = '301-500 тыс.'
df.loc[(df['BALLANCE'] >= 501000) & (df['BALLANCE'] <= 1000000) & (df['balance_bucket'].isna()), 'balance_bucket'] = '501-1 000 тыс.'
df.loc[(df['BALLANCE'] > 1000001) & (df['balance_bucket'].isna()), 'balance_bucket'] = '1 млн. и более'
```

```python
df.groupby('balance_bucket').agg({'BALLANCE' :'max'})
```

```python
df.groupby('balance_bucket').agg({'BALLANCE' :'min'})
```

```python
df.loc[(df['EXPIRE_DAYS'] <= 360) , 'expire_bucket'] = '360-'
df.loc[(df['EXPIRE_DAYS'] >= 361) & (df['EXPIRE_DAYS'] <= 540) & (df['expire_bucket'].isna()), 'expire_bucket'] = '361-540'
df.loc[(df['EXPIRE_DAYS'] >= 541) & (df['EXPIRE_DAYS'] <= 720) & (df['expire_bucket'].isna()), 'expire_bucket'] = '541-720'
df.loc[(df['EXPIRE_DAYS'] >= 721) & (df['EXPIRE_DAYS'] <= 900) & (df['expire_bucket'].isna()), 'expire_bucket'] = '721-900'
df.loc[(df['EXPIRE_DAYS'] >= 901) & (df['EXPIRE_DAYS'] <= 1080) & (df['expire_bucket'].isna()), 'expire_bucket'] = '901-1080'
df.loc[(df['EXPIRE_DAYS'] >= 1081) & (df['EXPIRE_DAYS'] <= 1260) & (df['expire_bucket'].isna()), 'expire_bucket'] = '1081-1260'
df.loc[(df['EXPIRE_DAYS'] >= 1261) , 'expire_bucket'] = '1261+'
```
