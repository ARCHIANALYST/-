```python
import pandas as pd
```

```python
current_contracts = pd.read_excel(r'C:\Users\Arthur\Documents\Data\rshb\application.xlsx', sheet_name = 'Действующие_контракты')
current_contracts
```

```python
print(current_contracts.isnull().sum())
print(current_contracts.info())
# знакомимся с датафреймом Действующие контракты, замечаем пропуски и некорректный тип данных
```

```python
current_contracts.dropna(how='all',inplace=True)
current_contracts.reset_index(drop=True,inplace=True)
# Избавляемся от пропусков
```

```python
current_contracts['superid'] = current_contracts['superid'].astype('int64')
current_contracts['client_number'] = current_contracts['client_number'].astype('int64')
# Корректиурем тип данных
```

```python
print(current_contracts.isnull().sum())
print(current_contracts.info())
current_contracts
# теперь все хорошо
```

```python
current_clients = pd.read_excel(r'C:\Users\Arthur\Documents\Data\rshb\application.xlsx', sheet_name = 'Клиенты')
current_clients
```

```python
print(current_clients.isnull().sum())
print(current_clients.info())
# знакомимся с датафреймом Клиенты, все ок
```

<b>1.</b> Выбрать уникальные кредитные договора за 01.01.2023<br>

```python
task_1 = current_contracts['agreement_number'][current_contracts['on_date'] == '2023-01-01'].unique()
task_1 = pd.DataFrame(data=task_1)
task_1.columns = ['agreement_number']
task_1.to_excel (r'C:\Users\Arthur\Documents\Data\rshb\task_1.xlsx', index= False)
task_1
```

<b>2.</b> Отобрать кредитные договора за 01.01.2023, сумма которых менее 400 000 и кредитная заявка заканчивается на 79 (кредитная заявка число)<br>

```python
task_2 = current_contracts.query('on_date == "2023-01-01" and limit < 400000 and superid.astype("str").str.endswith("79")')
task_2.reset_index(drop=True, inplace=True)
task_2 = task_2['agreement_number']
task_2 = pd.DataFrame(data=task_2)
task_2.to_excel (r'C:\Users\Arthur\Documents\Data\rshb\task_2.xlsx', index= False)
task_2
```

<b>3.</b> Вывести Номер заявки, договор и ФИО клиента, которые родился в Москве без дублей.<br>

```python
task_3 = current_contracts.merge(current_clients, on='client_number', how='inner')
task_3 = task_3[['superid', 'agreement_number', 'FIO']][task_3['place_birth'] == 'г. Москва'].reset_index(drop=True)
task_3.drop_duplicates(inplace=True)
task_3.to_excel (r'C:\Users\Arthur\Documents\Data\rshb\task_3.xlsx', index= False)
task_3
```

<b>4.</b> Вывести всю информацию с таблицы «Действующие контракты» с дополнительными полем – сумма всех кредитных договоров у клиента. Наименование можно дать любое

```python
task_4_group  = current_contracts.groupby(['client_number','on_date'])['limit'].sum()
task_4_group = task_4_group.reset_index()
task_4_group.columns = ['client_number','on_date', 'total_limit']
task_4_group
# task_4_group = task_4_group.reset_index()
# task_4_group.columns = ['client_number', 'total_limit']
# task_4_group
```

```python
task_4 = current_contracts.merge(task_4_group, on=['client_number', 'on_date'], how='inner')
task_4['on_date'] = task_4['on_date'].dt.strftime('%m-%d-%Y')
task_4['agreement_date'] = task_4['agreement_date'].dt.strftime('%m-%d-%Y')
task_4
```

```python
task_4.to_excel (r'C:\Users\Arthur\Documents\Data\rshb\task_4.xlsx', index= False)
task_4
```
